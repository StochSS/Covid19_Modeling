```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
```

```python
posterior_sb = np.loadtxt("results/SB_posterior.csv")
posterior_buncombe = np.loadtxt("results/Buncombe_posterior.csv")
```

```python
param_names = ['beta', 'kappa', 'delta', 'nu']
lower_bounds =  [0     ,       0,       0,    0]
upper_bounds =  [3     ,       1,     0.1,    1]
```

```python
fig, ax = plt.subplots(posterior_sb.shape[1],posterior_sb.shape[1])
for i in range(posterior_sb.shape[1]):
    for j in range(posterior_sb.shape[1]):
        if i > j:
            ax[i,j].axis('off')
        elif i == j:
            sns.distplot(posterior_sb[:,i], ax = ax[i,j], bins = 'auto')
            sns.distplot(posterior_buncombe[:,i], ax = ax[i,j], color = 'C1', bins = 'auto')
        else:
            ax[i,j].scatter(posterior_sb[:,j], posterior_sb[:,i])
            ax[i,j].scatter(posterior_buncombe[:,j], posterior_buncombe[:,i])
    ax[0,i].set_title(param_names[i], fontsize = 20)
fig.set_size_inches(20, 20)
```

```python
param_names = ['beta', 'delta']

fig, ax = plt.subplots(1,2)

for i in range(2):
    sns.distplot(posterior_sb[:, i], ax = ax[i], color = 'C0', label = 'SB', norm_hist = True)
    sns.distplot(posterior_buncombe[:,i], ax = ax[i], color = 'C1', label = 'Buncombe', norm_hist = True)
    ax[i].set_title(param_names[i], fontsize = 10)
    
    if i == 0:
        ax[i].legend()
fig.set_size_inches(20, 8)
```
